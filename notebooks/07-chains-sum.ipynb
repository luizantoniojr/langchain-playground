# Chains Sum

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
prompt = ChatPromptTemplate.from_template('Fale uma curiosidade sobre o seguinte assunto: {theme}')

chain_curiosity = prompt | ChatOpenAI() | StrOutputParser()
```

```python
prompt = ChatPromptTemplate.from_template('Crie uma história sobre o seguinte tema: {theme}')

chain_story = prompt | ChatOpenAI() | StrOutputParser()
```

```python
final_chain = chain_curiosity | chain_story

final_chain.invoke({"theme": "Marte"})
```
