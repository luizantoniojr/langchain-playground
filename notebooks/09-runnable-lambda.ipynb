# Runnable Lambda

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
from langchain_core.runnables import RunnableLambda

def greet(name):
    return f"Olá, {name}!"

greet_lambda = RunnableLambda(greet)

chain = greet_lambda | StrOutputParser()

chain.invoke("John")
```
