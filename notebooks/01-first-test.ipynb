# Meu primeiro teste com LangChain

### Iniciando LLM

```python
from langchain_openai import OpenAI

llm = OpenAI()
```

### Chamando LLM

```python
question = "Qual a capital do Brasil?"
for step in llm.stream(question):
    print(step, end="")
```

### Perguntas Simuntâneas

```python
questions = [
    "Qual é a capital de São Paulo?",
    "Qual é a capital de Minas Gerais?",
    "Qual é a capital de Rio de Janeiro?",
    "Qual é a capital de Pernambuco?",
]

llm.batch(questions)
```

### Chat Models

```python
from langchain_openai import ChatOpenAI

chat = ChatOpenAI()
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

messages = [
    SystemMessage(content="Você é um assistente de IA que responde perguntas sobre a capital do Brasil."),
    HumanMessage(content="O que você pode me dizer sobre a capital do Brasil?"),
]

for step in chat.stream(messages):
    print(step.content, end="")
```
